```python
def read_text(dict):
        file = open(dict,mode='rt',encoding="utf8")
        text = file.read()
        file.close()
        return text
```

```python
def to_lines(text):
      word = text.strip().split('"\n')
      
      word = [i.split('","') for i in word]
      
      
    
      return word
```

```python
import nltk
def tokenize_and_tag(sentence):
    tokens = nltk.word_tokenize(sentence)
    #print("Parts of speech:\n", nltk.pos_tag(tokens))
    return nltk.pos_tag(tokens)
```

```python
def normalize_pos(pos_tag):
    if pos_tag == 'WRB' or pos_tag == 'RB' or pos_tag == 'JJ':
        return 'Adverb'
    elif pos_tag == 'VBP' or pos_tag == 'VBZ' or pos_tag == 'VB' or pos_tag == 'VBN' or pos_tag == 'VBG' or pos_tag == 'VBD':
        return 'Verb'
    elif pos_tag == 'PRP' or pos_tag == 'PRP$' or pos_tag == 'WP':
        return 'Pronoun'
    elif pos_tag == 'NN' or pos_tag == 'NNP' or pos_tag == 'NNPS' or pos_tag == 'NNS':
        return 'Noun'
    elif pos_tag == 'DT' or pos_tag == 'CD' or pos_tag == 'WDT':
        return 'Determiner'
    elif pos_tag == 'IN' or pos_tag == 'TO':
        return 'Preposition'
```
